```python
import pandas as pd
```

```python
# read data
df = pd.read_csv("hpimonthlyandqtlytables1to19.csv", header=None)
df.head(20)
```

```python
# drop extra rows
df = df.drop([0,1])
df = df.reset_index(drop=True)
df
```

```python
# fill missing data
df[[4,6,8,10,12]] = df[[4,6,8,10,12]].bfill()
df[1] = df[1].ffill()
df = df.drop([0,2,3,5,7,9,11],axis=1)
df
```

```python
# convert data to numeric
for i in [4,6,8,10,12]:
    df[i] = pd.to_numeric(df[i].str.replace(',', ''))
    
df
```

```python
# Making data appropriate for visualization by data quartely intervals in column for year.
data_part_2 = df.copy()
# Iterate through the first column and add 0.25, 0.50, and 0.75 to every 1st,2nd and 3rd entry respectively
for i in range(0, len(data_part_2), 4):
    data_part_2.iloc[i:i+3, 0] += [0.25, 0.50, 0.75]

# Display the DataFrame
print(data_part_2)
```

```python
# Define the column names
column_names = ['Year', 'New Dwellings Price', 'Other Dwellings Price', 'All Dwellings Price', 'First time buyers Price', 'Former owner occupiers Price']

# Assign the column names to the DataFrame
data_part_2.columns = column_names

# Display the DataFrame
data_part_2
```

```python
# Save data for visualization 2
data_part_2.to_csv('London_house_prices_part2.csv',index=False)
```

```python
# Mean of every year price
df = df.groupby(1).mean()
df.reset_index(inplace=True)
df = df.astype(int)
df
```

```python
# Insert type of prices of each entry
df.insert(loc=1, column='Type1', value='New Dwellings Price')
df.insert(loc=3, column='Type2', value='Other Dwellings Price')
df.insert(loc=5, column='Type3', value='All Dwellings Price')
df.insert(loc=7, column='Type4', value='First time buyers Price')
df.insert(loc=9, column='Type5', value='Former owner occupiers Price')
df.columns
```

```python
# Flatten dataframe as per needs for chart
data = pd.DataFrame({
    'Year': df[1].repeat(5),
    'Type': df[['Type1','Type2','Type3','Type4','Type5']].values.flatten(),
    'Price': df[[4,6,8,10,12]].values.flatten()
})
data
```

```python
# Save data for visualization 2
data.to_csv('London_house_prices_part1.csv',index=False)
```
